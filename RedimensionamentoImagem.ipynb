```python
from PIL import Image
import matplotlib.pyplot as plt
image = Image.open("nature.jpg")

grayscale_image = image.convert("L")
grayscale_image.save("grayscale_nature.jpg")

threshold = 128  # Pixels above this value become white, below become black
bw_image = grayscale_image.point(lambda x: 255 if x > threshold else 0, mode="1")
bw_image.save("bw_nature.jpg")
```

```python
plt.imshow(grayscale_image, cmap="gray")
plt.axis("off")  # Remove eixos para uma visualização mais limpa
plt.show()
```

```python
plt.imshow(bw_image, cmap="gray")
plt.axis("off")
plt.show()
```
